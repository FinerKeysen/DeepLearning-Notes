```python
#!/usr/bin/python
# -*- coding: UTF-8 -*-
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('../demos/image001.jpg', cv2.IMREAD_COLOR) #BGR
plt.imshow(img)
plt.title('original image loaded from opencv is BGR')
plt.show()
```

```python
img_1 = img[::-1]
plt.imshow(img_1)
print "X"
plt.title('image[::-1]')
plt.show()
```

```python
img_2 = img[:, ::-1]
plt.imshow(img_2)
print "Y"
plt.title('image[:,::-1]')
plt.show()
```

```python
img_3 = img[:,:,::-1]
plt.imshow(img_3)
print "BGR->RGB"
plt.title('image[:,:,::-1]')
plt.show()
```
